## Standardize data
df.column_name.value_counts() show the frequency of the data in categories in the table

#### date

```python
df.date.value_counts()
```

```python
def convert_dmy(date):
    s1 = date.replace('-',' ').replace('/',' ').replace('.',' ')
    
    # replace all duplicated space and turn into one space
    s2 = " ".join(s1.split())
    d, m, y = s2.split(" ")
    if len(d) == 1:
        d = "0"+d
    else:
        pass

    if len(m) == 1:
        m = "0"+m
    else:
        pass

    if len(y) == 2:
        y = "20" + y
    else:
        pass
    return d,m,y

def standardize_date(df):
    for cell in df['date']:
        # & in cell means the string contains time
        if "&" in cell:
            date, time = cell.split('&')
            d, m, y = convert_dmy(date)
            cell = d + "/" + m + "/" + y
            
        else:
            d, m, y = convert_dmy(cell)
            cell = d + "/" + m + "/" + y
```

```python
standardize_date(df)
```

```python
# lower case the all values
df['job_location'] = df['job_location'].str.lower()
# repalce "chennai " to "chennai"
df.loc[df['job_location'] == 'chennai ', 'job_location'] = 'chennai'
# repalce "gurgaonr" to "gurgaon"
df.loc[df['job_location'] == 'gurgaonr', 'job_location'] = 'gurgaon'
# repalce "- cochin-" to "cochin"
df.loc[df['job_location'] == '- cochin- ', 'job_location'] = 'cochin'
```
