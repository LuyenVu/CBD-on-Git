### Information Gain of displacement:

Because displacement feature has maximum Information Gain ($Info_{displacement}=0$), so we pick displacement as the node for 2nd level without going on calculating other columns.

If __displacement__ = _low_ then __mpg__ is _good_.

If __displacement__ = _medium_ or _high_ then __mpg__ is _bad_.

The code for this decision tree using _If_ expression would be like this:

```python
def is_mpg_good(modelyear, displacement=""):
    if type(modelyear)==float or modelyear <70 or modelyear>83:
        print('modelyear must be an integer from 70 to 83')
        return
    if 70<=modelyear<=74:
        return 'bad'
    elif 75<=modelyear<=78:
        if displacement=="":
            raise TypeError("Please insert displacement status")
        elif displacement=="low":
            return 'good'
        elif displacement=="medium" or displacement=="high":
            return 'bad'    
        else:
            print('displacement must be a "low","medium" or "high"')
    elif 79<=modelyear<=83:
        return 'good'    
```

# Predict mpg

```python
yr = 70
print('model year = {} => mpg is'.format(yr), is_mpg_good(yr))

yr = 75
dpm = 'low'
print('model year = {}, displacement = {}, => mpg is'.format(yr,dpm), is_mpg_good(yr, dpm))

yr = 76
dpm = 'medium'
print('model year = {}, displacement = {}, => mpg is'.format(yr,dpm), is_mpg_good(yr, dpm))

yr = 77
dpm = 'high'
print('model year = {}, displacement = {}, => mpg is'.format(yr,dpm), is_mpg_good(yr, dpm))

yr = 80
print('model year = {} => mpg is'.format(yr), is_mpg_good(yr))
```

```python
hot = pd.get_dummies(df.iloc[:,1:].astype(str))
hot
```

```python
# convert input data from str to number
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X = df.iloc[:,1:].astype('str').apply(le.fit_transform)
y = df['mpg']
X
```

```python

# create DTs model
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(criterion="entropy")
tree.fit(hot, y)

from sklearn.tree import export_graphviz
import graphviz 
dot_data = export_graphviz(tree, out_file=None,
                          feature_names=hot.columns,
                          class_names=True,
                          filled=True, rounded=True,
                          special_characters=True,
#                           leaves_parallel=True,
                          )  
graph = graphviz.Source(dot_data)  
graph 
```

```python
from sklearn.metrics import classification_report
print('Classification_report: \n', 
      classification_report(
          y_true=y,
          y_pred=tree.predict(hot)))
tree.predict(hot.iloc[0].values.reshape(1,-1))
```
